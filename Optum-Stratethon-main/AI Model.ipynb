```python
import csv
reader = csv.reader(open(r'C:\Users\pskir\Downloads\6180e4c10462e_Case_Files\gapdays.csv', 'r'))
d = {}
for row in reader:
   k, v = row
   d[k] = v
```

```python
final_df["Days"] = [0]*1171
```

```python
row = -1
for x in final_df["Days"]:
    row+=1
    if final_df.loc[row,"Id"] in d.keys():
        final_df.loc[row,"Days"] = d[final_df.loc[row,"Id"]]
```

```python
from statistics import mean
```

```python
row=-1
for x in final_df["Medications"]:
    row += 1
    final_df.at[row,'Medications']=mean(x)
row=-1
for x in final_df["Procedures"]:
    row += 1
    final_df.at[row,'Procedures']=mean(x)
```

```python
final_df["Medications"]=(final_df["Medications"]-final_df["Medications"].mean())/final_df["Medications"].std()
final_df["Procedures"]=(final_df["Procedures"]-final_df["Procedures"].mean())/final_df["Procedures"].std()
```

```python
row = -1
for x in final_df["Days"]:
    row+=1
    final_df.at[row,'Days']=int(round(float(x)))
```

```python
convert_dict = {'Medications': float,
                'Procedures': float,
                'Days': int
               }
  
final_df = final_df.astype(convert_dict)
```
